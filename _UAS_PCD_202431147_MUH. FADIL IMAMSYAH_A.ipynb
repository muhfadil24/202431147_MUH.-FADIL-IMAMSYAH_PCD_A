## IMPORT LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## MEMBACA GAMBAR

```python
image = cv2.imread("fadil.jpeg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## MENAMPILKAN GAMBAR ASLI

```python
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.title("Citra Asli")
plt.axis("on")
plt.show()
```

## MEDIAN FILTERING

```python
median = cv2.medianBlur(image,5)
median_rgb = cv2.cvtColor(median,cv2.COLOR_BGR2RGB)
```

## MEAN FILTERING SECARA MANUAL

```python
def mean_filter_manual(img, kernel=3):

    tinggi, lebar = img.shape

    hasil = np.zeros_like(img)

    pad = kernel // 2

    padded = np.pad(img,
                    ((pad,pad),(pad,pad)),
                    mode='constant')

    for i in range(tinggi):
        for j in range(lebar):

            area = padded[i:i+kernel,
                          j:j+kernel]

            hasil[i,j] = np.mean(area)

    return hasil
```

## MENJALANKAN FUNGSI MEAN

```python
mean = mean_filter_manual(gray,3)
```

## MENAMPILKAN  HASIL I

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Citra Asli")
plt.axis("On")

plt.subplot(2,2,2)
plt.imshow(median_rgb)
plt.title("After Filter Median")
plt.axis("on")

plt.tight_layout()
plt.show()
```

## MENAMPILKAN HASIL II

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,3)
plt.imshow(gray,cmap='gray')
plt.title("Original Image")
plt.axis("on")

plt.subplot(2,2,4)
plt.imshow(mean,cmap='gray')
plt.title("Mean Filtered Image")
plt.axis("on")

plt.tight_layout()
plt.show()
```
